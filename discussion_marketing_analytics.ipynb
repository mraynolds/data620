Netflix and Marketing Analytics

Netflix has always been an innovator in the home entertainment market. Beginning as a mail-order DVD rental service, Netflix went on to pioneer and continually innovate the entertainment streaming service space, going quickly from a content provider to content producer. Netflix's efforts have made them a leading tech company and enshrined them in the acronym FAANG alongside the likes of Facebook, Amazon, Apple, and Google.

At the core of Netflix's business has always been customization and machine learning. Embedded in that is the use of analytics to customize almost every part of the user experience to maximize all aspects of a consumers relationship with the company.

The use of analytics including marketing analytics has been an integral and evolving part of their success with specific strategies and KPIs changing over time and, sometimes, quite rapidly. Due to the subscription nature of Netflix, there are generally three major strategies that the marketing department serves.

- Subscriber attraction: Attracting new or return subscribers
- Subscriber retention: Keeping active customers enrolled
- Maximizing ad-supported programming revenue

Below is a brief discussion of each of these strategies and Key Performance Indicators that either have been used by the company or could be an element of how they evaluate their performance:

Customer Attraction:

The building of the Netflix subscriber base year over year is done in a multifaceted and tiered way. Marketing is done both for the service as a whole and for specific productions or assets. This marketing runs the entire range of the marketing spectrum. This could be as simple as a poster at a bus stop, to elaborate film premieres for original productions, to developing viral moments and guerilla campaigns that leverages relationships with organizations like the NFL and WWE and youtubers like Mr. Beast. These promotions serve both attracting new subscribers and retaining existing subscribers.

A large part of the goal is to make the content the topic of conversation, to keep it present in the subscribers and potential subscribers mind and on their lips. Ultimately for customer attraction, the metric of new subscribers is paramount.

Customer  Retention:

As mentioned above, the general marketing campaigns serve as marketing for subscriber retention. Similar to the metric for subscriber attraction, the metric for subscriber retention is, on its face relatively simple. Of course, it can be framed in many ways that make it more complex, such as type of retention, whether the subscription type was changed during the time period in question, and more.

One of the primary ways marketing is used for existing subscribers is through the internal user interface and recommendation system. What Netflix offers to the subscriber to watch, how that is presented (different art images are displayed based on the subscriber), and more are all leveraged as personalized recommendations. This system is intended to keep the subscriber engaged and build the company's relationship to them. This is primarily done through machine learning algorithms. The recommendation system is a marketing tool used to reduce churn and increase lifteime value and engagement.

Additional metrics are valid for establishing performance when it comes to retaining subscribers. Some of the most common are number of views of a product, total hours a product is viewed, hours a subscriber uses the service, completion rate for a product (do they finish it or stop watching at a certain point), and many more.

Maximizing Ad-Supported Revenue:

Netflix has approached in-program advertising much the way it approaches customer retention, through highly customized programming based on specific user recommendations and knowledge gathered from subscriber habits and patterns. Knowledge of the subscriber allows for customized and targeted ad placement, with the goal of increasing traditional ad revenue metrics such as brand favorability, sales per impression, purchase intent, and more. Use of these metrics, and perhaps more importantly, continuing to vary offerings based on an AI based algorithm to refine what is being offered are the leading edges of Netflix's in-program advertising strategy.

Netflix is nothing if not analytics driven, while much of the structure of their advertising and strategic goals, such as subscriber attractions and retention are relatively traditional in nature, it is their constant refining of tactics through machine that drives their marketing.
