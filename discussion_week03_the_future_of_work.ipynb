Week Three - The Future of Work

Please read this essay:

“In the future, will there be any work left for people to do?,” Geoff Colvin, June 2nd 2014.

In this essay (and his more recent book, Humans are Underrated) Colvin talks about how “InfoTech” like text mining and analysis is disrupting the world of work.  An example cited in the book is that “e-discovery would enable one lawyer to do the work of 500 or more…”

Can you think of another example where text mining or network analysis might either eliminate jobs, allow one person to do the work of many, or perform new work that was not previously possible?

Please post your thoughts by Thursday, and respond to at least one classmate’s post by Sunday.

Feedback

In many ways, we have already seen the advancement of Geoff Colvin's thesis. Large language models have proliferated in the last several years and begun replacing or have replaced all sorts of roles. In addition to basic level lawyer work, much of basic level customer service has been replaced with automated chatbots that have the capacity to attempt to address a customer concern and rectify it. There are companies that are using LLMs to handle the first round of interviews as a form of screening, making the first actual human contact in the second round of interviewing.

There is definitely a feeling of sea change being underway in many industries. A feeling of the fundamental structures of how much of our world interacts is being reformulated and restructured through technology. An example from the company I work for of a role that I expect has a strong potential to be eliminated due to technological improvements in network analysis is the production planner. The role of the production planner is a central critical supply chain role that works to set SKUs to be produced based on product availability and sales, and then sets how they will be distributed regionally throughout the company based on individual production, supply and demand of the different production sites. It is a time intensive role that requires a small team to consider a lot of variables, and then make decisions about production and distribution and forecasts for the near future. It is easy to see how a well programmed algorithm with the right inputs could replace the entire team.

However, as Colvin indicates, this would not entirely eliminate the work needed to do these jobs, but it would reshape the work being done in a meaningful way. There would still need to be at least one employee or consultant (the team is currently 3, that work very long hours), to maintain and monitor the algorithm and ensure the development and the input of the correct data. In this way the work would be reshaped rather than eliminated.
