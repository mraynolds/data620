A code-based sentiment analysis to consider for a business use would be the following study  which considered the sentiment of Tweets about the personal finance app Robinhood in the period just before and the period just after Robinhood stopped allowing individual purchases of GameStop stock in 2021. This action was precipitated by an incident involved a group of individual traders, primarily organized around a reddit subgroup, attempting to short squeeze GameStop stock.

https://www.r-bloggers.com/2021/04/analyzing-brand-sentiment-with-robinhood-gamestop-r/

1. The business domain being discussed in this analysis is statements related to the public perception of Robinhood on Twitter. In the abstract, this is a brand sentiment analysis of informal communication preceding and following a well publicized incident. This is a particularly interesting form of brand feedback as it is not necessarily directed at the brand in the same way a public review would be, and it is not necessarily made by customers but instead is public discourse and sentiment about the brand.
2. The specific problem this sentiment analysis addresses is how does public perception of a brand change after a major brand action, possibly under adverse conditions. More generally it addresses brand sentiment before and after a major change of some type. An interesting sub-element of the analysis is: how long does that perception change last. Within this analysis the brand perception was already making a return to the mean after a week. So a secondary possible analysis is: what is the half life/lifespan of a change in perception? Are they durable or limited?
3. This analysis performed the sentiment analysis using r language and the packages tidyverse and tidytext. Both the afin and bing sentiment libraries were applied in order to apply a sentiment score to each analyzed word. Then an analysis of the proportion of all words and their sentiment was charted over time showing the shift in the sentiment of tweets over time following the cessation of GameStop trading.
4. This approach is widely applicable as a tool in multiple facets. It can be used to monitor the sentiment of a brand in overall online discourse, it can be used to monitor the shift in sentiment of the discourse of a brand following a major event such as a product release, a product change, a product elimination, or another major brand related news event. Finally, the sheer volume of words and tweets could be used to monitor overall engagement around the brand. This widely applicable use should be explored by brands to see in which ways constant monitoring may be leveraged or informative about what strategic and tactical decisions to make.
