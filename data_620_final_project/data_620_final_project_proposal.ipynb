### Final Project Proposal


#### Predicting whether a film review receives a New York Times' Critic's Pick designation based on textual and network-based features.

The New York Times reviews many films every year. Films that are particularly well received by the reviewing critic are given a designation as a "Critic's Pick".

This project will work to classify whether a film received a "Critic's Pick" designation utilizing textual and network-based features and work to answer whether the designation can be predicted from the text of the review and what effect the centrality measure of actors and directors might have on the prediction.

The data will be acquired through the New York Times API. An adequate amount of reviews will be acquired to create a meaningful network and corpus with which to work. The expectation that this will be a minimum of 5 to 8 years and as much as several decades.

A bi-partite network will be built of film titles and the directors and the top billed actors. The network will be limited to the largest connected component. A projected network of actors and directors will be made and centrality measures calculated.

A potential subsection of the overall analysis will be as follows. (It is considered a subsection as it does not necessarily contribute directly to answering the main question. As the data will have been acquired, the analysis will be potentially interesting): A projected network of films will be made and centrality measure calculated. This will then be graphed and marked with films that received Critic's Picks to see if they tend to cluster or not.

The text of the reviews will be tokenized and processed using TF-IDF potentially as unigrams, bigrams, and trigrams.

The processed text and the centrality measures will be assembled as features and used to train classifiers. Features will then be selected and the classifiers will be tuned and the results be evaluated. Then the trained models will be applied to a reserved test set of reviews to evaluate the classifiers performance. There may be a comparison of the classifier's performance with and without centrality measurements.
