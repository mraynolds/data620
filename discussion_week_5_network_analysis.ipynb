Network analytics are a powerful tool for fraud detection. There will always be inherent explicit and implicit connections between people and groups that commit fraud. A network allows those connections to be drawn in a way that might not otherwise have been apparent.

A classic example is bank fraud: Fraudsters open accounts, build credit, borrow and then maximize their accounts and disappear. Often, one way that could be identified is that they will utilize the same legitimate data in many iterations in order to maximize the number of fraudulent accounts they create. It is the reuse of data, that presents an opportunity to identify the fraudulent accounts even before the fraud occurs.

Creating and analyzing a network of the data will help identify these accounts. Things like repeat addresses, names, social security numbers, phone numbers, all have the potential to be reused and to help identify the connection between fraudsters and accounts. A network walk can begin to help identify adjacent nodes. Centrality meausures can also help identify who or what might be central to a subgraph of data.  Perhaps a single phone number is connected to many accounts. This is readily identifiable in network analysis. Clustering can then help identify related accounts and attributes that may be relevant. Finally, the aspects identified in this subgraph can then be used to search for similarly structured subgraphs that may be entirely unrelated but have a comparable structure.

Error rate is an important consideration. If fraud is 1% of all transactions, and a model to detect fraud fails everytime, it will look like it is 99% successful and has only 1% error rate. Of course this is not a success. Additionally, a model that yields a lot of false positive will cause chaos and confusion for clients and be costly in other ways. Therefore, a model generally needs to be balance precision and recall, catching the majority of the fraud.
