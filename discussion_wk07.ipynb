Describe an interesting use case that combines network data analysis and text processing.  Include the business goals, and the available or potential sources of data.  [What you describe here might evolve into a candidate topic for your final project!]

 Online reviews are a ubiquitous part of existence in the modern world. Products, restaurants, events, museums, parks:  everything gets a review these days. The democratization of media has created the democratization of criticism. Many prominent websites exist entirely to review things: Yelp, Trip Advisor, Google Business Profiles on Maps, Angi, Facebook reviews. Oftentimes, small businesses may live and die on their review rating.

These reviews provide a corpus to analyze and an inherent social network of reviewers connected by things they reviewed. In many instances this data can be accessed through an API or can be web-scraped, either directly or via third party scraping platform.

With this collection of data and its network there are some very interesting use cases that can be pursued:

A common issue in review culture is less than truthful reviews. This can take several forms:

- Fake reviews promoting a business:
        Through analysis it may be possible to identify fake reviews for a business. Through analyzing the sentiment of the review, the location of the business related to the reviewer, the amount of reviews related to the age of the business, it may be possible to identify suspicious reviews. This will then allow for tracing whether other reviews by the same reviewer might also be false. This would help identify what businesses might be padding their reviews, falsifying reviews or hiring reviewers, and who might be reviewing. From a business perspective it may be useful for a review site to have a way of analyzing who may be bad actors both amongst their listed businesses and their users.

- Review bombing:
         Part of modern internet culture allows for a business to be inundated with often negative (but sometimes positive) reviews. This is often based on a viral moment of some kind that incites people to review the business despite never having been there. A network analysis and language analysis might identify reviews created in this environment and if there are groups of people who are particularly prone to engaging in this type of action. Again, this is a place a site who hosts revieww might want to identify where reviews might essentially be spam and automatically negate them or de-emphasize them.

 Perhaps, considering the above, an interesting business use case would be a reviewer rating, either as an independent site or as an internal metric at the review host site. For an independent site, a review rating might tell what review rating could be trusted based on analysis of things like fake reviews, review bombing, or even the average sentiment of a user. Similarly, a site might want to create an internal rating for a reviewer based on their overall sentiment, possible veracity, and centrality. This would allow the business to decide whether a review should be empahsized or de-emphasized.
