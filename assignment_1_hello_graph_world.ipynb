```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
adjacency_dict = {'Andre': ('Beverly', 'Carol', 'Diane', 'Fernando'), 'Beverly' : ('Diane', 'Garth', 'Ed'), 'Carol': ('Fernando', 'Diane'), 'Diane': ('Fernando', 'Garth', 'Ed'), 'Ed':('Garth',), 'Fernando': ('Garth','Heather'), 'Garth' :('Heather',), 'Heather':('Ike',),'Ike':('Jane',)}
kite = nx.Graph(adjacency_dict)
kite.edges()
```

```python
pos = {
    'Andre':(0,0),
    'Beverly':(0,-1.5),
    'Carol':(1,0.75),
    'Diane':(1,-0.75),
    'Ed':(1,-2.25),
    'Fernando':(2,0),
    'Garth':(2,-1.5),
    'Heather':(3,-0.75),
    'Ike':(4,-0.75),
    'Jane':(5,-0.75)
}

label_pos = {}
labels = {}

for k, v in pos.items():
    label_pos[k] = (v[0]+0.15,v[1]-0.15)

for k in pos:
    labels[k] = k

plt.figure(figsize=(10, 10))

ax = plt.gca()

nx.draw(kite, pos, ax=ax, with_labels=False, width = 4.0, edge_color = 'purple')
nx.draw_networkx_labels(kite,label_pos,labels)
ax.set_aspect('equal')
ax.axis('off')
```
