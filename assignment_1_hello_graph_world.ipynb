```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.image as mpimg
```

```python
# create the network relationships

adjacency_dict = {'Andre': ('Beverly', 'Carol', 'Diane', 'Fernando'), 'Beverly' : ('Diane', 'Garth', 'Ed'), 'Carol': ('Fernando', 'Diane'), 'Diane': ('Fernando', 'Garth', 'Ed'), 'Ed':('Garth',), 'Fernando': ('Garth','Heather'), 'Garth' :('Heather',), 'Heather':('Ike',),'Ike':('Jane',)}

# create the networkx graph

kite = nx.Graph(adjacency_dict)

# check the network edges

kite.edges()
```

```python
# set positions in networkx so that the graph is consistent on every render

pos = {
    'Andre':(0,0),
    'Beverly':(0,-1.5),
    'Carol':(1,0.75),
    'Diane':(1,-0.75),
    'Ed':(1,-2.25),
    'Fernando':(2,0),
    'Garth':(2,-1.5),
    'Heather':(3,-0.75),
    'Ike':(4,-0.75),
    'Jane':(5,-0.75)
}

# define dictionaries for positioning labels below the nodes instead of on them

label_pos = {}
labels = {}

for k, v in pos.items():
    label_pos[k] = (v[0]-0.1,v[1]-0.15)

for k in pos:
    labels[k] = k

# load the node image into python
img = mpimg.imread('assignment1image.jpg')

# create the figure object and axes

fig, ax = plt.subplots(figsize=(8,8))


# add node image to plot

for node, (x,y) in pos.items():
    imagebox = OffsetImage(img, zoom=0.15)
    ab = AnnotationBbox(imagebox, (x,y), frameon=False)
    ax.add_artist(ab)


# draw network edges and place labels
nx.draw_networkx_edges(kite, pos, ax=ax,  width = 4.0, edge_color = 'purple')
nx.draw_networkx_labels(kite,label_pos,labels, font_weight='bold', horizontalalignment='left')
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```
