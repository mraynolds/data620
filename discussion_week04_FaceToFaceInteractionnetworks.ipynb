An historical example of a figure who was effective at hacking/navigating their social network was Harriet Tubman. After escaping slavery herself, she led some 13 missions to rescue approximately 70 enslaved people. She also went on to become an armed scout and spy for the Union army during the civil war. The ability to lead so many missions and help free so many people along the underground railroad seems a pinnacle example of navigating a social network. In fact, while her existing link would have been valuable, there is an extent to which her ability to form new links in the network would make an analysis of the network over time very interesting.

In all, betweenness centrality would be an interesting measure to consider, especially over time as the network changed as new links were formed. Also, with her role in direct action, analyzing the eigenvector centrality would be interesting, not necessarily because she would be the most central figure but because it would give insight into how that type of direct action role appears within the decentralized network of churches, abolitionist organizations, and individuals that comprised the Underground Railroad.
