Import necessary packages and name list.

```python
import nltk
import random
from nltk.corpus import names
```

Create a list of labeled names.

```python
labeled_names = ([(name, 'male') for name in names.words('male.txt')] + [(name, 'female') for name in names.words('female.txt')])
```

Shuffle the name list.

```python
random.seed(1234)
random.shuffle(labeled_names)
```

#### Possible variables in a name:

To create a name gender predictor, it will be necessary to enter features that will aid in a model being more successful at predicting the gender commonly associated with a name. In order to execute this task, it is worthwhile to narrow down possible variables that an algorithm could use. Below is a list of possible variables:

- Name length
- Starting letters (1 to n letters from the beginning of the name)
- Ending letters (1 to n letters from the end of the name)
- Specific letters
- Count of letters
- Specific position of letters
- Specific grouping of consonants (as in do specific sets of consonants play a role)

While this list is not exhaustive, it covers many possibilities.

#### Testing the first variable

```python
def gender_features(name):
    features = {}
    features['name_length'] = len(name)

    return features
```
